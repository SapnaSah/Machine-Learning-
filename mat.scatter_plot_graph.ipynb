```python
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
plt.scatter(day,no,color="g")
plt.show()
```

```python
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
no2=[12,13,19,41,25,33,36]                                                #two graph in one plot.
colors=[20,30,40,60,80,90,67]
sizes=[140,260,280,100,320,135,248]
plt.scatter(day,no,c=colors,s=sizes,cmap="viridis")
plt.scatter(day,no2,c=colors,s=sizes)                                    #two graph in one plot.
t=plt.colorbar()
t.set_label("colorbar",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("scatter plot",fontsize=15)
plt.show()
```
