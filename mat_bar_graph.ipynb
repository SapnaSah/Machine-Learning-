```python
import matplotlib.pyplot as plt
import numpy as np
x=["python","c","c++","java"]
y=[85,70,60,82]
z=[20,30,40,50]    
p=np.arange(len(x))
plt.xlabel("language",fontsize=15)
plt.ylabel("language",fontsize=15)
plt.title("bargraph",fontsize=15)
width=0.2


plt.bar(p,y,width,color="r",label="popular")   
plt.bar(p,z,width,color="y",label="popular1")      
plt.show()
```

```python
x=["python","c","c++","java"]
y=[85,70,60,82]
z=[20,30,40,50]  

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]                              # z graph mean yellow graph ko bagal krne ke liye.

plt.xlabel("language",fontsize=15)
plt.ylabel("language",fontsize=15)
plt.title("bargraph",fontsize=15)



plt.bar(p,y,width,color="r",label="popular")   
plt.bar(p1,z,width,color="y",label="popular1")
plt.show()
```

```python
x=["python","c","c++","java"]
y=[85,70,60,82]
z=[20,30,40,50]  

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("language",fontsize=15)
plt.ylabel("language",fontsize=15)
plt.title("bargraph",fontsize=15)



plt.bar(p,y,width,color="r",label="popular")   
plt.bar(p1,z,width,color="y",label="popular1")
plt.xticks(p+width/2,x)                                           #it is used to  name like pyhon etc.                            
plt.show()
```

```python
x=["python","c","c++","java"]
y=[85,70,60,82]
z=[20,30,40,50]  

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("language",fontsize=15)
plt.ylabel("language",fontsize=15)
plt.title("bargraph",fontsize=15)



plt.bar(p,y,width,color="r",label="popular")   
plt.bar(p1,z,width,color="y",label="popular1")
plt.xticks(p+width/2,x,rotation=10)                   #rotaion used for name like java python  ectc are tiled.
plt.show()
```
