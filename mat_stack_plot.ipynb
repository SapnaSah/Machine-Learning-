```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
area=[1,2,3,4,5]
plt.stackplot(x,area)
plt.show()
```

```python
x=[1,2,3,4,5]
area=[2,3,2,5,4]
plt.stackplot(x,area)
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[2,3,2,5,4]
area2=[2,4,6,7,10]
area3=[3,2,4,5,6]
plt.stackplot(x,area1,area2,area3)
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[2,3,2,5,4]
area2=[2,4,6,7,10]
area3=[3,2,4,5,6]
l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l)
plt.legend()
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[2,3,2,5,4]
area2=[2,4,6,7,10]
area3=[3,2,4,5,6]
c=["y","m","r"]
l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l,colors=c)
plt.legend()
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[2,3,2,5,4]
area2=[2,4,6,7,10]
area3=[3,2,4,5,6]
c=["y","m","r"]
l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l,colors=c,baseline="sym")
plt.legend()
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[2,3,2,5,4]
area2=[2,4,6,7,10]
area3=[3,2,4,5,6]
c=["y","m","r"]
l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l,colors=c,baseline="wiggle")
plt.legend()
plt.show()
```
