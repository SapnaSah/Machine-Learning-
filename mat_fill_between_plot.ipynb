```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,3,4,6,7,9])
y=np.array([1,3,4,6,7,9])
plt.plot(x,y,color="m")
plt.fill_between(x,y,color="r",where=(x>=2) & (x<=6))

plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,3,4,6,7,9])
y=np.array([1,3,4,6,7,9])
plt.plot(x,y,color="m")
plt.fill_between(x,y,color="r",where=(x>=2) & (x<=6),alpha=0.4)

plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
