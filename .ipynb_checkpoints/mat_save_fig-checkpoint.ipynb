```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,3,6,5]
plt.plot(x,y,color="y")                                                        #  only plot uses krne se linear graph banta 
plt.savefig("line")                                                            # savefig ka use krke ham graph save kr skte h isi folder ke andar.                                   
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,3,6,5]
plt.plot(x,y,color="y")                                                       
plt.savefig("line1",dpi=2000)                                              # dpi means dot per inch jo quality of dot hota h                                      

plt.show()                                                                    
```

```python
x=[1,2,3,4,5]
y=[2,4,3,6,5]
plt.plot(x,y,color="y")                                                        
plt.savefig("line1",dpi=2000,facecolor="y")                                                                                    

plt.show()    
```

```python
x=[1,2,3,4,5]
y=[2,4,3,6,5]
plt.plot(x,y,color="y")                                                       
plt.savefig("line3.pdf",dpi=2000)                                                                                     

plt.show()    
```

```python
x=[1,2,3,4,5]
y=[2,4,3,6,5]
plt.plot(x,y,color="y")                                             
plt.savefig("line123.jpg",dpi=2000)                                                                                     

plt.show()    
```

```python
x=[1,2,3,4,5]
y=[2,4,3,6,5]
plt.plot(x,y,color="y")                                                        #  only plot uses krne se linear graph banta 
plt.savefig("line123.jpg",dpi=2000,transparent=True)                                                  # savefig ka use krke ham graph save kr skte h isi folder ke andar.                                   

plt.show()    
```
