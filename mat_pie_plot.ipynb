```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["c","c++","java","python"]
plt.pie(x)
plt.show()
```

```python
x=[10,20,30,40]
y=["c","c++","java","python"]
plt.pie(x,labels=y)
plt.show()
```

```python
x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.4,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex)                 #it used to look different of all.
plt.show()
```

```python
x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.0,0.0,0.2,0.0]
plt.pie(x,labels=y,explode=ex)                 #it used to look different of all.
plt.show()
```

```python
x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.4,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%")                #0.1f means one 0 show after point.          
plt.show()
```

```python
x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.4,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex,autopct="%0.2f%%")                #0.2f means two zero(0) show after point.          
plt.show()
```

```python
x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.4,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex,autopct="%0.2f%%",shadow="true")                # shadow is used to show thw like 3d images.         
plt.show()
```

```python
x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.4,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex,autopct="%0.2f%%",shadow="true",radius=2.0)                # radius is used to increase the size.         
plt.show()
```

```python
x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.4,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex,autopct="%0.2f%%",shadow="true",labeldistance=1.4)         # labeldistance is used  to the  distance between graph and its names.         
plt.show()
```

```python
x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.4,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex,autopct="%0.2f%%",shadow="true",labeldistance=1.1)              
plt.show()
```

```python
x=[10,20,30,40]
y=["c","c++","java","python"]                                                    
ex=[0.4,0.0,0.0,0.0]                                                 #startangle to used the rotate angle.
plt.pie(x,labels=y,explode=ex,autopct="%0.2f%%",shadow="true",labeldistance=1.4,startangle=90)               
plt.show()
```

```python
x=[10,20,30,40]
y=["c","c++","java","python"]                                             #textprop is used to the increse size of text means 10.00 etc.       
ex=[0.4,0.0,0.0,0.0]                                                
plt.pie(x,labels=y,explode=ex,autopct="%0.2f%%",radius=2,shadow="true",labeldistance=1.4,startangle=0,textprops={"fontsize":15})               
plt.show()
```

```python
x=[10,20,30,40]
y=["c","c++","java","python"]                                             #textprop is used to the increse size of text.       
ex=[0.4,0.0,0.0,0.0]                                                
plt.pie(x,labels=y,explode=ex,autopct="%0.2f%%",radius=1.5,shadow="true",labeldistance=1.4,startangle=90,textprops={"fontsize":30})               
plt.show()
```

```python
x=[10,20,30,40]
y=["c","c++","java","python"]                                             # wedgeprops is used to increase the outer line shadow.     
ex=[0.4,0.0,0.0,0.0]                                                
plt.pie(x,labels=y,explode=ex,autopct="%0.2f%%",radius=1.5,shadow="true",labeldistance=1.4,startangle=0,textprops={"fontsize":14},counterclock="false"
       ,wedgeprops={"linewidth":4})               
plt.show()
```

```python
x=[10,20,30,40]
y=["c","c++","java","python"]                                             # edgecolor is used to  color the outer line shadow.     
ex=[0.4,0.0,0.0,0.0]                                                
plt.pie(x,labels=y,explode=ex,autopct="%0.2f%%",radius=1.5,shadow="true",labeldistance=1.4,startangle=0,textprops={"fontsize":14},counterclock="false"
       ,wedgeprops={"linewidth":4,"edgecolor":"m"})               
plt.show()
```

```python
x=[10,20,30,40]
y=["c","c++","java","python"]                                             #  rotatelabel is used to rotate the label.     
ex=[0.4,0.0,0.0,0.0]                                                
plt.pie(x,labels=y,explode=ex,autopct="%0.2f%%",radius=1.5,shadow="true",textprops={"fontsize":14},counterclock="false"
       ,wedgeprops={"linewidth":4,},rotatelabels=True)               
plt.show()
```

```python
x=[10,20,30,40]
y=["c","c++","java","python"]                                             #  rotatelabel is used to rotate the label.     
ex=[0.4,0.0,0.0,0.0]                                                
plt.pie(x,labels=y,explode=ex,autopct="%0.2f%%",radius=1.5,shadow="true",textprops={"fontsize":14},counterclock="false"
       ,wedgeprops={"linewidth":4,},rotatelabels=False)   

plt.show()
```

```python
x=[10,20,30,40]
y=["c","c++","java","python"]                                                 
ex=[0.4,0.0,0.0,0.0]                                                
plt.pie(x,labels=y,explode=ex,autopct="%0.2f%%",radius=1,shadow="true",textprops={"fontsize":14},counterclock="false"
       ,wedgeprops={"linewidth":4,},rotatelabels=False)   
plt.title("pie_graph")

plt.show()
```

```python
x=[10,20,30,40]
y=["c","c++","java","python"]                                           
ex=[0.4,0.0,0.0,0.0]                                                
plt.pie(x,labels=y,explode=ex,autopct="%0.2f%%",radius=1,shadow="true",textprops={"fontsize":14},counterclock="false"
       ,wedgeprops={"linewidth":4,},rotatelabels=False)   
plt.title("pie_graph")
plt.legend()
plt.show()
```

```python
x=[10,20,30,40]
y=["c","c++","java","python"]                                                     # location the graph present.
ex=[0.4,0.0,0.0,0.0]                                                
plt.pie(x,labels=y,explode=ex,autopct="%0.2f%%",radius=1,shadow="true",textprops={"fontsize":14},counterclock="false"
       ,wedgeprops={"linewidth":4,},rotatelabels=False)   
plt.title("pie_graph")
plt.legend(loc=1)
plt.show()
```

```python
#multiple pie chart/donar pie chart.
x=[10,20,30,40]
x1=[15,25,35,45]
c=["y","m","b","g"]
y=["c","c++","java","python"]
plt.pie(x,labels=y,radius=1.5)
plt.pie(x1,radius=1,colors=c)
plt.show()
```

```python
x=[10,20,30,40]
x1=[15,25,35,45]
c=["y","m","b","g"]
y=["c","c++","java","python"]
plt.pie(x,labels=y,radius=1.5)
plt.pie([1],colors="w")
plt.show()
```
