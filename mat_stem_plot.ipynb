```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[2,2,5,4,4,3]
plt.stem(x,y)
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,4,4,3]
plt.stem(x,y,linefmt=":")
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,4,4,3]
plt.stem(x,y,linefmt="--")
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,4,4,3]
plt.stem(x,y,linefmt="-.")
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,4,4,3]
plt.stem(x,y,linefmt=":",markerfmt="y")
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,4,4,3]
plt.stem(x,y,linefmt=":",markerfmt="y*")
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,4,4,3]      
plt.stem(x,y,linefmt=":",markerfmt="yo")             #wapas dot point ke liye o ka use krte h.
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,4,4,3]
plt.stem(x,y,linefmt=":",markerfmt="y*",bottom=1)
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,4,4,3]
plt.stem(x,y,linefmt=":",markerfmt="y*",bottom=9)
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,4,4,3]
plt.stem(x,y,linefmt=":",markerfmt="y*",bottom=0,basefmt="m")
plt.show()
```

```python
x=[1,2,3,4,5,6]                                                                   #use_line_collection=False.
y=[2,2,5,4,4,3]
plt.stem(x,y,linefmt=":",markerfmt="yo",bottom=0,basefmt="m",label="python",orientation="horizontal")
plt.legend()
plt.show()
```
