```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60]
plt.boxplot(x)
plt.show()
```

```python
x=[10,20,30,40,50,60]
plt.boxplot(x,notch=True)                                 #notch is used to change yhe shape.
plt.show()
```

```python
x=[10,20,30,40,50,60]
plt.boxplot(x,vert=False)
plt.show()
```

```python
x=[10,20,30,40,50,60]
plt.boxplot(x,widths=2)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60]
plt.boxplot(x,labels=["python"])
plt.show()
```

```python
x=[10,20,30,40,50,60]
plt.boxplot(x,labels=["python"],patch_artist=True)                            # fill the color
plt.show()
```

```python
x=[10,20,30,40,50,60]
plt.boxplot(x,labels=["python"],patch_artist=True,showmeans=True)                     # show the means value.
plt.show()
```

```python
x=[10,20,30,40,50,60,130]                       #jab data limit se bahar jata h to outler ho jata h.
plt.boxplot(x,labels=["python"])
plt.show()
```
