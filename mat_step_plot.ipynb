```python
import matplotlib.pyplot as plt
x=[1,3,5,7,9]
y=[10,20,35,56,60]


plt.step(x,y)
plt.show()
```

```python
x=[1,3,5,7,9]
y=[10,20,35,56,60]

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.step(x,y)
plt.show()
```

```python
x=[1,3,5,7,9]
y=[10,20,35,56,60]

# plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.step(x,y)
plt.show()
```

```python
x=[1,3,5,7,9]
y=[10,20,35,56,60]

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.step(x,y)
plt.show()
```

```python
x=[1,3,5,7,9]
y=[10,20,35,56,60]

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.step(x,y,marker="o")
plt.show()
```

```python
x=[1,3,5,7,9]
y=[10,20,35,56,60]

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.step(x,y,marker="o",ms=15)
plt.show()
```

```python
x=[1,3,5,7,9]
y=[10,20,35,56,60]

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.step(x,y,marker="o",ms=15,color="m")                           #marker sise is ms.
plt.show()
```

```python
x=[1,3,5,7,9]
y=[10,20,35,56,60]

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.step(x,y,marker="o",ms=15,color="m",mfc="g")                    #marker face color.
plt.show()
```

```python
x=[1,3,5,7,9]
y=[10,20,35,56,60]

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.step(x,y,marker="o",ms=15,color="m",mfc="g",label="python")
plt.legend()
plt.show()
```
